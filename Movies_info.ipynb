```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
csv_file = 'data/Movies.csv'
Movies_info_df = pd.read_csv(csv_file)
Movies_info_df.head()
```

```python
Movies_filtered_df = Movies_info_df.drop(Movies_info_df.index[[0,4,5,6,7]])
Movies_filtered_df
```

```python
Movies_filtered_df['Name'] = Movies_filtered_df.Name.str.title()
Movies_filtered_df
```

```python
engine = create_engine("postgresql://postgres:XXXXXXX@localhost:5432/ETL_ProjectDB")
conn = engine.connect()

Movies_filtered_df.to_sql(name='movies', con=engine, if_exists='append', index=False)


```
